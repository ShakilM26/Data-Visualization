```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crash = sns.load_dataset('car_crashes')
```

```python
# PairPlot

tips = sns.load_dataset('tips')
sns.pairplot(tips, hue='sex', palette='Blues')
```

```python
# Rugplot

sns.displot(tips['tip'])
sns.rugplot(tips['tip'])
```

```python
sns.set_style('white')
sns.set_context('paper', font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol', data=crash, kind='reg')
sns.despine(left=True)
```

```python
sns.set_style('dark')
sns.stripplot(x='day', y='total_bill', data=tips, hue='sex')
sns.swarmplot(x='day', y='total_bill', data=tips, color='white')
plt.legend(loc=0)
```

```python
# Matrix / Heatplot

sns.set_context('paper', font_scale=1.4)
corr = crash.corr()
corr

sns.heatmap(corr, annot=True)
# Annot showing you the number
```

```python
# Heatmap

fli = sns.load_dataset('flights')
pivot = fli.pivot_table(index='month', columns='year', values='passengers')

sns.heatmap(pivot, cmap='Blues', linecolor='white', linewidth=0.5)
```

```python
# Cluster map

iris = sns.load_dataset('iris')
species = iris.pop('species')
sns.clustermap(iris)
# Try with flights
```
