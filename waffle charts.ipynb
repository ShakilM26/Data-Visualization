* Waffle Chart is an alternative way to Pie Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
```

```python
lith = {'Lith':['Shale','Sandstone','Sandstone/Shale','Chalk',
               'Limestone', 'Marl', 'Tuff'],
       'Well1':[61,15,10,5,5,3,1],
       'Well2':[35,21,16,12,7,5,4]}
lith_df = pd.DataFrame(lith)
lith_df
```

```python
colours = ['#8dd3c7', '#deb887', '#bebada', '#fb8072',
           '#80b1d3', '#fdb462', '#b3de69']
plot_labels = [f'{i} ({str(j)} %)' for i,j in zip(lith_df.Lith,
lith_df.Well1)]
```

```python
plt.figure(FigureClass=Waffle, figsize=(10,10), rows=5, columns=20,
          values=list(lith_df['Well1']),
          colors=colours, labels=plot_labels,
          legend={'loc':'lower center', 'bbox_to_anchor':(0.5,-0.8),
                 'ncol':3, 'fontsize':12})
plt.show()
```

```python
fig = plt.figure(FigureClass=Waffle, 
                 plots = {211: {'values':list(lith_df['Well1']),
                               'labels': [f'{i} ({str(j)} %)' for i,j in zip(lith_df.Lith, 
                                                    lith_df.Well1)],
                               'legend':{'loc':'center left', 'bbox_to_anchor': (1.0, 0.5), 
                                          'ncol':1, 'fontsize':12},
                                'title':{'label':'Well 1 Lithology Composition', 'fontsize':18}
                               },
                               
                          212: {
                              'values':list(lith_df['Well2']),
                              'labels': [f'{i} ({str(j)} %)' for i,j in zip(lith_df.Lith, 
                                         lith_df.Well2)],
                              'legend':{'loc':'center left', 'bbox_to_anchor': (1.0, 0.5), 
                                          'ncol':1, 'fontsize':12},
                              'title':{'label':'Well 2 Lithology Composition', 'fontsize':18}
                          }
                         },
                 figsize=(15,10), 
                 rows=5, 
                 columns = 20, 
                 colors=colours)
                 
plt.tight_layout()
plt.show()
```
