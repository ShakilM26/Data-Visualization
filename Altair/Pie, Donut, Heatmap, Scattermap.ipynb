```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('world_population 2020.csv')
df.head()
```

```python
Types = ['Continent','Yearly_Change','Density(P/Km²)','World_Share','Fertility_Rate']
avg = df[Types].groupby(by='Continent').mean().T.reset_index().rename(columns={'index':'Types'})
avg = avg[:]
avg
```

# Pie Chart

```python
alt.Chart(avg).mark_arc().encode(theta=alt.Theta(field='Asia', type='quantitative'),
                                color=alt.Color(field='Types', type='nominal'),
                                tooltip=['Types', 'Asia']).properties(width=400, title='Avg. Calculation of Per Continent')
```

# Donut Chart

```python
alt.Chart(avg).mark_arc(innerRadius=80).encode(theta=alt.Theta(field='Asia', type='quantitative'),
                                color=alt.Color(field='Types', type='nominal'),
                                tooltip=['Types', 'Asia']).properties(width=400, title='Avg. Calculation of Per Continent')
```

# Radial Chart

```python
alt.Chart(avg).mark_arc(innerRadius=20).encode(theta=alt.Theta(field='Asia', type='quantitative'),
                                               radius=alt.Radius('Asia', scale=alt.Scale(type='sqrt', zero=True)),
                                color=alt.Color(field='Types', type='nominal'),
                                tooltip=['Types', 'Asia']).properties(width=400, title='Avg. Calculation of Per Continent')
```

```python
alt.Chart(avg).mark_arc(innerRadius=20).encode(theta=alt.Theta('Asia', stack=True),
                                               radius=alt.Radius('Asia', scale=alt.Scale(type='sqrt', zero=True)),
                                color=alt.Color('Types'),
                                tooltip=['Types', 'Asia']).properties(width=400, title='Avg. Calculation of Per Continent')
```
