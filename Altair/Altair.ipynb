```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('student performance.csv')
df.head()
```

```python
# bar chart 

alt.Chart(df).mark_bar().encode(x='race/ethnicity', y='writing score', color='race/ethnicity').properties(height=200, width=400)
```

```python
# horizontal bar chart with tooltip

alt.Chart(df).mark_bar().encode(y='race/ethnicity', x='math score', color='race/ethnicity', tooltip=[alt.Tooltip('reading score'),
                                                                                                     alt.Tooltip('writing score')]).properties(height=200, width=350)
```

```python
from vega_datasets import data 

df = data.cars()
df.head()
```

```python
# scatter plot using point with different uses of tooltip

alt.Chart(df).mark_point().encode(x='Horsepower', y='Miles_per_Gallon', color='Origin', tooltip=['Name','Cylinders'])
```

```python
# see the difference ? 
# I fill this point with 'filled' parameter

alt.Chart(df).mark_point(filled=True, size=50).encode(x='Horsepower', y='Miles_per_Gallon', color='Origin', tooltip=['Name','Cylinders'])
```

```python
# scatter Circle

alt.Chart(df).mark_circle(size=55).encode(x='Acceleration', y='Miles_per_Gallon', color='Origin', tooltip=['Name','Horsepower'])
```

```python
df = pd.read_csv('diamonds.csv')
df.head()
```

```python
alt.Chart(df).mark_bar().encode(x='cut', y='carat', color='cut').properties(height=200, width=300, title='Dimonds Grading Scale')
```

```python
alt.Chart(df, width=500, title='Cut Distribution').mark_bar(color='blue').encode(
    x=alt.X('cut',title='Cut Type', sort=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'],
            axis=alt.AxisConfig(labelAngle=45)),
            y=alt.Y('count()', title='Observation Count'),
            color=alt.Color('clarity'))
```

```python
# Histogram 

alt.Chart(df, title='Carat Distribution in Histogram').mark_bar(color='#1AE9CA').encode(alt.X('carat', bin=alt.Bin(maxbins=20), axis=alt.Axis(titleFontSize=15)),
                                               alt.Y('count()', axis=alt.Axis(titleFontSize=15))).configure_axis(grid=False, domain=True)
```

```python
alt.Chart(df).mark_boxplot().encode(x='carat')

# this circle are outliers value
```
