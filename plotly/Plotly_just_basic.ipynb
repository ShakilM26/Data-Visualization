```python
import pandas as pd
import plotly.express as pe
import plotly.graph_objects as go
```

```python
df = pe.data.gapminder()
print(df.columns)
df['continent'].unique()
```

**Bar chart**

```python
pe.bar(df, x='continent', y='gdpPercap', color='continent')
```

**Histogram**

```python
pe.histogram(df, x='continent', y='pop', color='continent')
```

**scatter & bubble**

```python
df = pe.data.tips()
df.columns
```

```python
fig = pe.scatter(df, x='day', y='total_bill')
fig.update_layout(autosize=False)
```

```python
fig = pe.scatter(df, x='size', y='tip', size='tip', color='sex')
fig.update_layout(autosize=False)
```

**Pie Chart**

```python
pe.pie(df, values='tip', names='day')
```

**GANTT chart**

```python
# Generalized Activity Normalization Time Table

import plotly.figure_factory as ff

df = [dict(Task='A', Start='2020-02-21', Finish='2019-10-14'),
      dict(Task='B', Start='2020-04-06', Finish='2020-11-11'),
      dict(Task='C', Start='2021-10-10', Finish='2022-01-02')]
fig=ff.create_gantt(df)
fig.show()
```

**Error Bar's**

```python
df = pe.data.iris()
df['error'] = df['petal_length']/100

fig = pe.scatter(df, x='species', y='petal_width', error_x='error', error_y='error')
fig.show()
```
