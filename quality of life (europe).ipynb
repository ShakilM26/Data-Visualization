```python
import sweetviz as sv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('europe.csv')
df.head()
```

```python
columns = ['Quality of Life Index', 'Cost of Living Index', 'Safety Index', 'Health Care Index', 'Climate Index']

for i in columns:
    df[i] = df[i].astype(float)

plt.figure(figsize=(10, 8))
cnt = 1
for i in columns:
    plt.subplot(4, 2, cnt)
    sns.histplot(df[i], color='orange', kde_kws={'linewidth':10})
    plt.title(i)
    cnt = cnt+1
plt.tight_layout()
plt.show()
```

```python
corr = df.corr()

fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(corr, annot=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```
